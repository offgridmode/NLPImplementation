```python
import pandas as pd
```

```python
messages = pd.read_csv('spam.csv', sep=',', names=["label", "message"],encoding = 'latin-1',usecols=[0,1])
```

```python
messages
```

```python
messages = messages.drop(0,)
```

```python
messages
```

```python
## DATA CLEANING and Preprocessing
import re ##regular expression library
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
stop_words = set(stopwords.words('english'))
corpus = []
for i in range(len(messages)):
        review = re.sub('[^a-zA-Z]',' ', messages['message'].iloc[i])
        review = review.lower()
        review = review.split()
        review = [ps.stem(word) for word in review if word not in stop_words]
        review = ' '.join(review)
        corpus.append(review)
```

```python
##Create the Bag of Words Model With N-Grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,ngram_range=(1,2)) #this makes it a binary BOW
```

```python
X=cv.fit_transform(corpus).toarray()
X
```

```python
y = pd.get_dummies(messages['label'])
y=y.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.20,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB().fit(X,y)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
