```python
import pandas as pd 
messages = pd.read_csv('spam.csv', sep= ',', names = ['label', 'message'], encoding= 'latin-1', usecols=(0,1))
```

```python
messages
```

```python
messages = messages.drop(0,)
messages
```

```python
import re
```

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'].iloc[i])
    review = review.lower()
    review = review.split()

    review = [lemma.lemmatize(word) for word in review]
    review = ' '.join(review)
    corpus.append(review)
```

```python
import nltk
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess  ##important
```

```python
## simple_preprocess converts a document into a list of lower case tokens ignoring tokens thats are too short or too long.
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
           words.append(simple_preprocess(sent))
```

```python
import gensim
```

```python
## train word2vec from scratch
model = gensim.models.Word2Vec(words, vector_size=100)
```

```python
## to get all the vocabulary 
model.wv.index_to_key
```
